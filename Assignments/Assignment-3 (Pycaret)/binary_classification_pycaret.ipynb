```python
# import pycaret
import pycaret
pycaret.__version__
```

```python
# import data
from pycaret.datasets import get_data
d=get_data('heart_disease')
```

```python
from pycaret.classification import *
setup_ = setup(d, target='Disease', session_id=123)
```

```python
# import ClassificationExperiment and init the class
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
# init setup on exp
exp.setup(d, target='Disease', session_id=123)
```

```python
# compare models
best_model = compare_models()
```

```python
# evaluate model
evaluate_model(best_model)
```

```python
holdout_pred = predict_model(best_model)
```

```python
holdout_pred.head()
```

```python
new_data = d.copy()
new_data.drop('Disease', axis=1, inplace=True)
new_data.head()
```

```python
predictions = predict_model(best_model, data = new_data)
predictions.head()
```
